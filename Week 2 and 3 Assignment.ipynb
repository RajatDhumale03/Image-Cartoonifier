Part B

```python
def compute_cost(number_of_year_experience, salary, w, b):
  predicted_salary = w * number_of_year_experience + b
  total_cost = (1 / (2 * 3)) * np.sum((predicted_salary - salary)**2)
  return total_cost
```

Part C

```python
def compute_gradient(number_of_year_experience, salary, w, b):
  predicted_salary = w * number_of_year_experience + b
  djdw = (1 / 3) * np.sum((predicted_salary - salary) * number_of_year_experience)
  djdb = (1 / 3) * np.sum(predicted_salary - salary)
  return djdw ,djdb
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
dataset = data = pd.read_csv('/content/drive/MyDrive/corrected train set.csv')
test_dataset = data = pd.read_csv('/content/drive/MyDrive/week 1 test.csv')
```

```python
# Remove rows with missing target, separate target from predictors
#No need of - dataset.dropna(axis=0, subset=['failure'], inplace=True) as we don't have any missing values in target
y = dataset.failure
dataset.drop(['failure'], axis=1, inplace=True)
```

```python
# The dtype-'object' columns are - Product_code,attribute_0,attribute_1
# We can drop these columns as they will not have any significant effect
data = dataset.select_dtypes(exclude=['object'])
data_test = test_dataset.select_dtypes(exclude=['object'])
```

```python
data_train, data_valid, y_train, y_valid = train_test_split(data, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

# Function for calculating the mean absolute error
def score_dataset(data_train, data_valid, y_train, y_valid):
    model = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=4)
    model.fit(data_train, y_train)
    preds = model.predict(data_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
from sklearn.impute import SimpleImputer

#imputation
my_imputer = SimpleImputer()
imputed_data_train = pd.DataFrame(my_imputer.fit_transform(data_train))
imputed_data_valid = pd.DataFrame(my_imputer.transform(data_valid))

#imputation removed column names; put them back
imputed_data_train.columns = data_train.columns
imputed_data_valid.columns = data_valid.columns
```

```python
print(score_dataset(imputed_data_train, imputed_data_valid, y_train, y_valid))
```

```python
# Preprocessed training and validation features
final_imputer = SimpleImputer(strategy='median')
final_data_train = pd.DataFrame(final_imputer.fit_transform(data_train))
final_data_valid = pd.DataFrame(final_imputer.transform(data_valid))
#imputation removed column names; put them back
final_data_train.columns = data_train.columns
final_data_valid.columns = data_valid.columns
```

```python
# Define and fit model
model = XGBRegressor(n_estimators=1000, learning_rate=0.05, n_jobs=4)
model.fit(final_data_train, y_train)

# Get validation predictions and MAE
preds_valid = model.predict(final_data_valid)
print("MAE:")
print(mean_absolute_error(y_valid, preds_valid))
```

```python
#preprocess test data
final_data_test = pd.DataFrame(final_imputer.transform(data_test))
#imputation removed column names; put them back
final_data_test.columns = data_test.columns

#get test predictions
preds_test = model.predict(final_data_test)
```
